```python
2. What are the three different types of Boolean operators?
Answer:-
    
i.AND Operator(and)- the result will be True if both conditions are True.
Ex- if 5>2 and 0<1:
        print("both true")
    else:
        print("one is false")
O/P:- both true
    
ii. OR Operator(or) - the result will be True if any one condition is True
Ex- if 5>2 or 1>10:
        print("one is true")
    else:
        print("both are false")

iii. NOT Operator(not)- the result will be False if condition os True and vice versa.
Ex- if not 5<2:
        print("Condition is True because of not")
    else:
        print("Condition is False because of not")
```
