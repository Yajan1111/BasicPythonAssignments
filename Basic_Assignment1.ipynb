```python
3. Describe three different data types.
Answer:- 
    
Integers-int- This datatype stores whole numbers both negative and positive.
ex- 45,-79,5678 etc.
a=45
here a stores int value.

Floating Number- float- This datatype stores decimal values both negative & positive.
Ex- 45.5,3.14,-87.99 etc.
ex-
b=45.5
here b stores float value.

Strings-str-This datatype stores character stream inside '' or "".
Ex-"abc",'xyz','a',"gg","y","hello" etc.
c="qwerty"
here c stores str value.

Boolean-bool- This datatype stores either False or True.

ex- d= True
here d stores bool value
```
