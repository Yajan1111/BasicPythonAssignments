```python
#2. Why is PyInputPlus commonly imported with import pyinputplus as pypi?
Answer:-
Because pypi acts as alias to pyinputplus.So while calling built-in functions of pyinputplus we can call it as pypi.inputNum() rather than pyinputplus.inputNum() 
```

```python
#3. How do you distinguish between inputInt() and inputFloat()?
Answer:-
When you call inputInt() you are prompted to enter integer number.If you enter float number then error is thrown and ask you to enter again
When you call inputFloat() you are prompted to enter floating number.If you enter int number then it makes it to floating number.
```

```python
#4. Using PyInputPlus, how do you ensure that the user enters a whole number between 0 and 99?
Answer:-
pypi.inputInt(min=0,max=99)
```

```python
#5. What is transferred to the keyword arguments allowRegexes and blockRegexes?
Answer:-
We pass string to allowRegexes or blockRegexes if we want some values to be accepted and some values to get blocked while taking input.
```
