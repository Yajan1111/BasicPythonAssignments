```python
13. What's the difference between tuples and lists?
Answer:-
1. list is denoted by [] while tuple is denoted by ()
2. list is mutable while tuple is immutable
3. list has more built in methods while tuple has less
4. list consume more memory while tuple consume less
5. list is more useful when we need insertion or deletion while for tuple we can use it when it's read only or accessing it only once
```
